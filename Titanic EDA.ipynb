```python
df_train['Survived']= df_train['Survived'].map({1:'Yes',0:'No'})
```

```python
df_train.head()
```

```python
sns.histplot(x=df_train["Age"], label="Age",bins=10, kde=True)
plt.xlabel('Age')
plt.ylabel('Numbers')
plt.title("Survival Count by Age")
plt.legend()
plt.show()
```

```python
sns.histplot(x=df_train["Sex"], label="Gender",bins=10, kde=True)
plt.xlabel('Gender')
plt.ylabel('Numbers')
plt.title("Survival Count by Gender")
plt.legend()
plt.show()
```

```python
sns.catplot(x='Survived',col='Sex',kind='count',data=df_train)

sns.catplot(x='Survived',col='Embarked',kind='count',data=df_train)

sns.catplot(x='Survived',col="Pclass",kind='count',data=df_train)
```

```python
sns.catplot(x='Survived',col='SibSp',kind='count',data=df_train)

sns.catplot(x='Survived',col='Parch',kind='count',data=df_train)
```

```python
list1= sns.FacetGrid(df_train, hue='Survived', height=5)
list1.map(sns.histplot, 'Parch', kde=False, bins=10)
list1.add_legend()
plt.title('FacetGrid of Parch with Survived')
plt.show()

list2= sns.FacetGrid(df_train, hue='Survived', height=5)
list2.map(sns.histplot, 'SibSp', kde=False, bins=10)
list2.add_legend()
plt.title('FacetGrid of SibSp with Survived')
plt.show()
```

```python
list1= sns.FacetGrid(df_train, hue='Survived', height=5)
list1.map(sns.histplot, 'Age', kde=False, bins=10)
list1.add_legend()
plt.title('FacetGrid of Age with Survived')
plt.show()

list1= sns.FacetGrid(df_train, hue='Survived', height=5)
list1.map(sns.histplot, 'Sex', kde=False, bins=10)
list1.add_legend()
plt.title('FacetGrid of Sex with Survived')
plt.show()
```

```python
#BoxPlot

sns.boxplot(x="Survived", y = "Age", data= df_train)
plt.xlabel('Numbers')
plt.ylabel('Age')
plt.title("Survival Count by Age")
plt.legend()
plt.show()

sns.boxplot(x="Survived", y = "Parch", data= df_train)
plt.xlabel('Numbers')
plt.ylabel('Parch')
plt.title("Survival Count by Parch")
plt.legend()
plt.show()

sns.boxplot(x="Survived", y = "Fare", data= df_train)
plt.xlabel('Numbers')
plt.ylabel('Fare')
plt.title("Survival Count by Fare")
plt.legend()
plt.show()




```

```python
#Violin Plot

sns.violinplot(x='Survived', y = 'Age', data=df_train)
plt.title("Age")



```

```python
sns.violinplot(x='Survived', y = 'Fare', data=df_train)
plt.title("Fare")
```
